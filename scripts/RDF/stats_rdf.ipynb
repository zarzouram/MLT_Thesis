# Simple RDF

```python
rdf_nopq_df = rdf_df[(rdf_df['qualifier.value'].isna())
                     & (~rdf_df['object.value'].isna())]
```

```python
rdf_nopq_df = rdf_nopq_df.drop([
    'unit.value', 'unitLabel_ar.value', 'unitLabel_en.value',
    'object_sub.value', 'qualifier.value', 'pqtype.value',
    'qualifierLabel_ar.value', 'qualifierLabel_en.value',
    'object_subLabel_ar.value', 'object_subLabel_en.value',
    'object_sub.xml:lang', 'object.xml:lang', 'object_sub.datatype'
],
                               axis='columns')
```

```python
rdf_nopq_df.drop_duplicates(subset=[
    'subject.value', 'subjectLabel_ar.value', 'subjectLabel_en.value',
    'predicate.value', 'object.value', 'objectLabel_en.value',
    'objectLabel_ar.value', 'ptype.value', 'predicateLabel_ar.value',
    'predicateLabel_en.value'
],
                            inplace=True)
```

```python
display(rdf_nopq_df)
```

```python
rdf_nopq_df['object.datatype'].unique()
```

```python
rdf_nopq_df = rdf_nopq_df[(~rdf_nopq_df['objectLabel_en.value'].isna())
                          & (~rdf_nopq_df['objectLabel_ar.value'].isna())]

```

```python
print(f'There are {rdf_nopq_df.shape[0]:,} simple RDF; have no qualifiers')
```

## Level 0

```python
objects_ids = rdf_nopq_df['object.value'].unique()

rdf_nopq_level0_df = rdf_nopq_df.copy(
)[~rdf_nopq_df['subject.value'].isin(objects_ids)]
```

```python
print(
    f'There are {rdf_nopq_level0_df.shape[0]:,} simple RDF; have no qualifiers and level 0'
)
```

```python
subjects_num = rdf_nopq_level0_df['subject.value'].unique().shape[0]
predicte_num = rdf_nopq_level0_df['predicate.value'].unique().shape[0]
print(subjects_num, predicte_num)
```

```python
rdf_nopq_level0_df.groupby(
    by=['predicate.value'])['subject.value'].unique().apply(len).sort_values(
        ascending=False)
```
